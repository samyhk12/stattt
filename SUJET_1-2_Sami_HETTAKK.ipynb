## 1.1 Par inversion de la fonction de reÃÅpartition

### 1. G√©n√©ration de variables al√©atoires par inversion de la fonction de r√©partition

Soit $ U \sim \mathcal{U}[0,1] $, une variable al√©atoire uniform√©ment distribu√©e sur l‚Äôintervalle $ [0,1] $, et soit $ G $ une fonction croissante et bijective.  

L‚Äôobjectif est de d√©terminer l'expression de $ P(G(U) \leq y) $.

---

#### üîπ √âtape 1 : Expression de $ P(G(U) \leq y) $

√âtant donn√© que $ G $ est croissante et bijective, elle poss√®de une fonction r√©ciproque $ G^{-1} $. L‚Äôin√©galit√© $ G(U) \leq y $ est √©quivalente √† :  

$$
U \leq G^{-1}(y)
$$

Ainsi, la probabilit√© recherch√©e est donn√©e par :

$$
P(G(U) \leq y) = P(U \leq G^{-1}(y))
$$

Or, $ U $ suit une loi uniforme sur $ [0,1] $, ce qui implique :  

$$
P(U \leq t) = t, \quad \text{pour } t \in [0,1]
$$

Donc, on obtient :

$$
P(G(U) \leq y) = G^{-1}(y)
$$

---

#### üîπ √âtape 2 : Condition pour que $ Y = G(U) $ suive la loi de $ F_Y $

La fonction de r√©partition de $ Y $ est d√©finie par :  

$$
F_Y(y) = P(Y \leq y) = P(G(U) \leq y)
$$

D'apr√®s l'expression obtenue pr√©c√©demment :

$$
F_Y(y) = G^{-1}(y)
$$

Pour que $ Y $ suive la loi de r√©partition $ F_Y $, cette √©galit√© doit √™tre v√©rifi√©e pour tout $ y $. On en d√©duit que :

$$
G^{-1} = F_Y
$$

En prenant la fonction r√©ciproque des deux c√¥t√©s :

$$
G = F_Y^{-1}
$$

---

#### üéØ Conclusion  

La variable al√©atoire $ Y = G(U) $ suit la loi de r√©partition $ F_Y $ si et seulement si :

$$
G = F_Y^{-1}
$$

---

### 2. G√©n√©rer une variable al√©atoire suivant la loi exponentielle $E(\lambda)$

La loi exponentielle de moyenne $\frac{1}{\lambda}$, not√©e $E(\lambda)$, a pour fonction de r√©partition :

$$
F_Y(y) = P(Y \leq y) = 1 - e^{-\lambda y}, \quad y \geq 0
$$

Et la **fonction de r√©partition compl√©mentaire** (c'est-√†-dire la probabilit√© $P(Y \geq y)$) est :

$$
F_Y^C(y) = P(Y \geq y) = e^{-\lambda y}, \quad y \geq 0
$$

### Fonction indicatrice

On peut utiliser la **fonction indicatrice** pour exprimer cette fonction de r√©partition compl√©mentaire de mani√®re conditionnelle. La fonction indicatrice est d√©finie comme suit :

- $ \mathbb{1}_{\mathbb{R}^+}(y) = 1 $ si $ y \geq 0 $, sinon $ 0 $
- $ \mathbb{1}_{\mathbb{R}^-}(y) = 1 $ si $ y < 0 $, sinon $ 0 $

Donc, la fonction de r√©partition compl√©mentaire $F_Y^C(y)$ peut s'√©crire :

$$
F_Y^C(y) = P(Y \geq y) = e^{-\lambda y} \mathbb{1}_{\mathbb{R}^+}(y) + \mathbb{1}_{\mathbb{R}^-}(y)
$$

Cela signifie que la fonction de r√©partition compl√©mentaire est d√©finie de mani√®re explicite pour $ y \geq 0 $ avec la fonction indicatrice pour tenir compte du fait que $ Y $ prend uniquement des valeurs positives (puisque la loi exponentielle est d√©finie sur $ \mathbb{R}^+ $).

### M√©thode d'inversion de la fonction de r√©partition

La m√©thode d'inversion consiste √† g√©n√©rer une variable al√©atoire $ U \sim \mathcal{U}[0, 1] $ et √† appliquer la transformation inverse de la fonction de r√©partition pour obtenir une variable $ Y $ suivant la loi exponentielle. Pour la loi exponentielle, la fonction de r√©partition est :

$$
F_Y(y) = 1 - e^{-\lambda y}
$$

Il faut r√©soudre cette √©quation pour $ y $ en fonction de $ U $ :

$$
U = 1 - e^{-\lambda y}
$$

$$
e^{-\lambda y} = 1 - U
$$

$$
-\lambda y = \ln(1 - U)
$$

$$
y = -\frac{1}{\lambda} \ln(1 - U)
$$

### G√©n√©ration de la variable al√©atoire

Nous pouvons maintenant g√©n√©rer une variable al√©atoire $ Y $ suivant une loi exponentielle de param√®tre $ \lambda $ en utilisant la formule suivante :

$$
Y = -\frac{1}{\lambda} \ln(1 - U)
$$

Cependant, comme $ 1 - U $ suit la m√™me distribution uniforme que $ U $, on peut directement √©crire la formule comme suit :

$$
Y = -\frac{1}{\lambda} \ln(U)
$$

### Exemple avec $ \lambda = 2 $

Si $ \lambda = 2 $, la formule pour g√©n√©rer $ Y $ devient :

$$
Y = -\frac{1}{2} \ln(U)
$$

O√π $ U $ est une variable al√©atoire uniforme sur $ [0,1] $.

### Conclusion

La m√©thode d'inversion de la fonction de r√©partition permet de g√©n√©rer une variable al√©atoire $ Y $ suivant une loi exponentielle $ E(\lambda) $ en utilisant la transformation :

$$
Y = -\frac{1}{\lambda} \ln(U)
$$

Cela garantit que $ Y $ suit une loi exponentielle avec le param√®tre $ \lambda $, en prenant simplement une variable al√©atoire uniforme $ U \sim \mathcal{U}[0, 1] $ et en appliquant cette formule.

---


Impl√©mentation en Python :
Voici un code permettant de g√©n√©rer une variable al√©atoire Y suivant la loi exponentielle ùê∏(ùúÜ) avec Œª=2 :

```python
import numpy as np

# Param√®tre lambda pour la loi exponentielle
lambda_ = 2

# G√©n√©ration d'une variable al√©atoire suivant la loi uniforme U(0,1)
U = np.random.rand()

# Transformation inverse pour obtenir Y suivant E(lambda)
Y = -np.log(1 - U) / lambda_

# Affichage du r√©sultat
print(f"Valeur g√©n√©r√©e de Y : {Y}")
```

### 3. G√©n√©rer 10000 variables al√©atoires ind√©pendantes de loi $E(\lambda)$

La loi exponentielle $E(\lambda)$ de moyenne $\frac{1}{\lambda}$ et de variance $\frac{1}{\lambda^2}$ est d√©finie par sa fonction de r√©partition :

$$
F_Y(y) = 1 - e^{-\lambda y}, \quad y \geq 0
$$

Nous allons g√©n√©rer un √©chantillon de taille $I = 10000$ de variables al√©atoires ind√©pendantes suivant cette loi avec $\lambda = 2$. La m√©thode d'inversion permet de g√©n√©rer ces variables comme suit :

$$
Y = -\frac{1}{\lambda} \ln(U)
$$

o√π $U$ est une variable uniforme sur l'intervalle $[0,1]$.

### Expression analytique de la moyenne et de la variance

#### Moyenne de la loi exponentielle

La moyenne (ou esp√©rance) de la loi exponentielle $E(\lambda)$ est donn√©e par :

$$
\mathbb{E}[Y] = \frac{1}{\lambda}
$$

Pour $\lambda = 2$, la moyenne th√©orique est donc :

$$
\mathbb{E}[Y] = \frac{1}{2} = 0.5
$$

#### Variance de la loi exponentielle

La variance de la loi exponentielle $E(\lambda)$ est donn√©e par :

$$
\text{Var}(Y) = \frac{1}{\lambda^2}
$$

Pour $\lambda = 2$, la variance th√©orique est donc :

$$
\text{Var}(Y) = \frac{1}{2^2} = 0.25
$$

### Moyenne et variance empiriques

#### G√©n√©ration des variables al√©atoires

En Python, nous pouvons utiliser la m√©thode d'inversion pour g√©n√©rer les variables al√©atoires. Voici le code pour g√©n√©rer l'√©chantillon et calculer la moyenne et la variance empiriques :

```python
import numpy as np
import matplotlib.pyplot as plt

# Param√®tre lambda
lambda_val = 2
I = 10000

# G√©n√©ration des variables al√©atoires exponentielles
U = np.random.uniform(0, 1, I)
Y = -np.log(U) / lambda_val

# Moyenne et variance empiriques
mean_empirical = np.mean(Y)
variance_empirical = np.var(Y)

# Affichage des r√©sultats
print("Moyenne empirique :", mean_empirical)
print("Variance empirique :", variance_empirical)

# Histogramme de l'√©chantillon
plt.hist(Y, bins=50, density=True, alpha=0.6, color='g')
plt.title(f"Histogramme des variables exponentielles (lambda={lambda_val})")
plt.xlabel('Y')
plt.ylabel('Densit√©')
plt.show()
```

Cela montre que la m√©thode de g√©n√©ration fonctionne correctement, et les r√©sultats sont coh√©rents avec les propri√©t√©s th√©oriques de la loi exponentielle

### 4. V√©rifier la d√©croissance exponentielle de la fonction de r√©partition compl√©mentaire

La fonction de r√©partition compl√©mentaire pour la loi exponentielle $E(\lambda)$ est donn√©e par :

$$
F_Y^C(y) = P(Y \geq y) = e^{-\lambda y}, \quad y \geq 0
$$

Nous devons v√©rifier que l'√©chantillon g√©n√©r√© suit cette fonction de r√©partition compl√©mentaire.

#### √âtapes √† suivre :

1. **G√©n√©rer un √©chantillon de 10 000 variables al√©atoires ind√©pendantes** suivant la loi exponentielle $E(\lambda)$ avec $\lambda = 2$.
2. **Calculer la fonction de r√©partition compl√©mentaire empirique** $P(Y \geq y)$ pour les diff√©rentes valeurs de $y$ dans l'√©chantillon.
3. **Tracer la fonction de r√©partition compl√©mentaire** en √©chelle semilogarithmique, c'est-√†-dire en repr√©sentant $ \ln(1 - F_Y^C(y)) $ sur l'axe des ordonn√©es et $y$ sur l'axe des abscisses.
4. **V√©rifier que la pente de la droite obtenue** est proche de $-\lambda$.

#### Code Python pour g√©n√©rer l'√©chantillon et v√©rifier la d√©croissance exponentielle :


```python

import numpy as np
import matplotlib.pyplot as plt

# Param√®tre lambda
lambda_val = 2
I = 10000

# G√©n√©ration des variables al√©atoires exponentielles
U = np.random.uniform(0, 1, I)
Y = -np.log(U) / lambda_val

# Tri des donn√©es par ordre croissant
Y_sorted = np.sort(Y)

# Calcul de la fonction de r√©partition compl√©mentaire empirique : P(Y >= y)
F_c_empirical = np.arange(1, I + 1) / I

# Tracer de la fonction de r√©partition compl√©mentaire en √©chelle semilogarithmique
plt.figure(figsize=(8, 6))
plt.plot(Y_sorted, np.log(1 - F_c_empirical), label="Donn√©es exp√©rimentales", color="b")
plt.xlabel('y')
plt.ylabel('ln(1 - F_c(y))')
plt.title(f'Verification de la d√©croissance exponentielle (lambda={lambda_val})')
plt.grid(True)
plt.legend()
plt.show()

# Calcul de la pente √† partir des donn√©es
# On effectue un ajustement lin√©aire sur les valeurs en √©chelle semilogarithmique
from scipy.stats import linregress

slope, intercept, _, _, _ = linregress(Y_sorted, np.log(1 - F_c_empirical))

# Affichage de la pente
print(f"La pente de la droite est approximativement : {slope:.4f}")
```

L'erreur que je rencontre ("divide by zero encountered in log") se produit lorsque j'essaie de calculer le logarithme de 0, ce qui est math√©matiquement ind√©fini. Cette situation peut survenir lorsque les valeurs dans $1 - F_Y^C(y)$ deviennent exactement √©gales √† 0, ce qui arrive quand $F_Y^C(y)$ atteint exactement 1.

Pour r√©soudre cela, je peux filtrer ou ajuster les valeurs proches de 1 avant d'appliquer le logarithme, ou remplacer les valeurs trop proches de 1 par un petit nombre pour √©viter le probl√®me. Voici une modification de mon code pour √©viter cette erreur :

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Param√®tre lambda
lambda_val = 2
I = 10000

# G√©n√©ration des variables al√©atoires exponentielles
U = np.random.uniform(0, 1, I)
Y = -np.log(U) / lambda_val

# Tri des donn√©es par ordre croissant
Y_sorted = np.sort(Y)

# Calcul de la fonction de r√©partition compl√©mentaire empirique : P(Y >= y)
F_c_empirical = np.arange(1, I + 1) / I

# √âviter le logarithme de z√©ro en ajustant les valeurs proches de 1
epsilon = 1e-10  # Tr√®s petite valeur pour √©viter log(0)
F_c_empirical_adjusted = np.clip(F_c_empirical, epsilon, 1 - epsilon)

# Tracer de la fonction de r√©partition compl√©mentaire en √©chelle semilogarithmique
plt.figure(figsize=(8, 6))
plt.plot(Y_sorted, np.log(1 - F_c_empirical_adjusted), label="Donn√©es exp√©rimentales", color="b")
plt.xlabel('y')
plt.ylabel('ln(1 - F_c(y))')
plt.title(f'Verification de la d√©croissance exponentielle (lambda={lambda_val})')
plt.grid(True)
plt.legend()
plt.show()

# Calcul de la pente √† partir des donn√©es
slope, intercept, _, _, _ = linregress(Y_sorted, np.log(1 - F_c_empirical_adjusted))

# Affichage de la pente
print(f"La pente de la droite est approximativement : {slope:.4f}")
```

Le r√©sultat que j'ai obtenu pour la pente $-1.9924$ est tr√®s proche de $-2$, ce qui est coh√©rent avec la th√©orie. En effet, la pente de la droite, qui repr√©sente la relation

$$
\ln(1 - F_c(y)) = -\lambda y
$$

devrait √™tre proche de $-\lambda$, ici $\lambda = 2$.

### Interpr√©tation :
- J'ai g√©n√©r√© un √©chantillon de 10 000 variables al√©atoires de loi exponentielle avec $\lambda = 2$.
- En tra√ßant la fonction de r√©partition compl√©mentaire sur une √©chelle semilogarithmique, j'ai obtenu une droite dont la pente est tr√®s proche de $-\lambda$, ce qui confirme que la d√©croissance de la fonction suit bien la loi exponentielle avec le param√®tre $\lambda$ attendu.
- La l√©g√®re diff√©rence avec la valeur exacte $-2$ peut √™tre due √† des erreurs d'arrondi ou des petites fluctuations dans les donn√©es exp√©rimentales.

### Conclusion :
Mon analyse est correcte et la pente obtenue, $-1.9924$, est suffisamment proche de la valeur th√©orique $-2$, ce qui confirme que la fonction de r√©partition compl√©mentaire suit bien une d√©croissance exponentielle avec $\lambda = 2$.

#### **1. Question pr√©liminaire : Distribution de $Z = X_1^2 + X_2^2$**

Soit **$X_1$** et **$X_2$** deux variables al√©atoires ind√©pendantes suivant la loi normale centr√©e r√©duite **$\mathcal{N}(0,1)$**. On d√©finit la variable al√©atoire :

$$
Z = X_1^2 + X_2^2
$$

L'objectif est de montrer que **$Z$** suit une loi exponentielle de moyenne **$1/\lambda$** avec **$\lambda = \frac{1}{2}$**.

#### **Expression de $P(Z \geq z)$**
La probabilit√© cherch√©e est :

$$
P(Z \geq z) = P(X_1^2 + X_2^2 \geq z)
$$

En utilisant les densit√©s de **$X_1$** et **$X_2$**, cette probabilit√© peut s'√©crire sous la forme d'une int√©grale double :

$$
P(Z \geq z) = \iint_{x_1^2 + x_2^2 \geq z} \frac{1}{2\pi} e^{-\frac{x_1^2 + x_2^2}{2}} dx_1 dx_2
$$

#### **Changement de variable polaire**
On effectue le changement de variable :

$$
x_1 = r \cos\theta, \quad x_2 = r \sin\theta
$$

Le jacobien du changement de variable est **$J = r$**, donc :

$$
dx_1 dx_2 = r dr d\theta
$$

En coordonn√©es polaires, l'int√©grale devient :

$$
P(Z \geq z) = \int_0^{2\pi} \int_{\sqrt{z}}^{\infty} \frac{1}{2\pi} e^{-\frac{r^2}{2}} r dr d\theta
$$

#### **R√©solution de l'int√©grale**
On calcule d'abord l'int√©grale int√©rieure :

$$
\int_{\sqrt{z}}^{\infty} e^{-\frac{r^2}{2}} r dr
$$

En posant **$u = \frac{r^2}{2}$**, donc **$du = r dr$**, on obtient :

$$
\int_{\frac{z}{2}}^{\infty} e^{-u} du = e^{-\frac{z}{2}}
$$

Ainsi, l'int√©grale devient :

$$
P(Z \geq z) = \int_0^{2\pi} \frac{1}{2\pi} e^{-\frac{z}{2}} d\theta
$$

$$
= e^{-\frac{z}{2}} \int_0^{2\pi} \frac{1}{2\pi} d\theta
$$

$$
= e^{-2z}
$$

Finalement, la fonction de r√©partition compl√©mentaire de **$Z$** est :

$$
P(Z \geq z) = e^{-2z} \mathbb{1}_{\mathbb{R}^+}(z) + \mathbb{1}_{\mathbb{R}^-}(z)
$$

Ce qui montre que **$Z$** suit une loi exponentielle de param√®tre **$\lambda = 2$**, soit **$\mathcal{E}(2)$**.

#### **2. Distribution de $r^2$ et de $\theta$**

Soit $X_1$ et $X_2$ deux variables al√©atoires ind√©pendantes suivant une loi normale centr√©e r√©duite $\mathcal{N}(0,1)$. On d√©finit les coordonn√©es polaires :

$$
X_1 = r \cos\theta, \quad X_2 = r \sin\theta
$$

o√π $r$ est la distance √† l‚Äôorigine et $\theta$ est l‚Äôangle.

#### **1. Montrons que $r^2$ suit une loi exponentielle $\mathcal{E}(1/2)$**

D‚Äôapr√®s la question pr√©c√©dente, la somme des carr√©s des deux variables suit une loi exponentielle :

$$
Z = X_1^2 + X_2^2 = r^2 \sim \mathcal{E}(1/2)
$$

La fonction de r√©partition compl√©mentaire d‚Äôune loi exponentielle de param√®tre $\lambda$ est :

$$
P(Z \geq z) = e^{-\lambda z}
$$

Dans notre cas :

$$
P(r^2 \geq z) = e^{-z/2}
$$

Ce qui montre que $r^2$ suit une loi exponentielle de param√®tre $\frac{1}{2}$, soit :

$$
r^2 \sim \mathcal{E}(1/2)
$$

La moyenne de cette loi exponentielle est :

$$
\mathbb{E}[r^2] = \frac{1}{\lambda} = 2
$$

Ce qui confirme que $r^2 \sim \mathcal{E}(1/2)$.

---

#### **2. Montrons que $\theta$ suit une loi uniforme sur $[0,2\pi]$**

La densit√© jointe de $(X_1, X_2)$ est donn√©e par :

$$
f_{X_1, X_2}(x_1, x_2) = \frac{1}{2\pi} e^{-\frac{x_1^2 + x_2^2}{2}}
$$

En passant aux coordonn√©es polaires, le changement de variable $x_1 = r \cos\theta, x_2 = r \sin\theta$ donne :

$$
dx_1 dx_2 = r \, dr \, d\theta
$$

Ce qui entra√Æne la densit√© jointe en coordonn√©es polaires :

$$
f_{r, \theta}(r, \theta) = f_{X_1, X_2}(r \cos\theta, r \sin\theta) \times |J|
$$

Avec $J = r$, on trouve :

$$
f_{r, \theta}(r, \theta) = \frac{1}{2\pi} e^{-\frac{r^2}{2}} r
$$

On int√®gre par rapport √† $r$ pour obtenir la densit√© marginale de $\theta$ :

$$
f_{\theta}(\theta) = \int_0^\infty f_{r, \theta}(r, \theta) \, dr
$$

$$
= \int_0^\infty \frac{1}{2\pi} e^{-\frac{r^2}{2}} r \, dr
$$

Le terme $\int_0^\infty e^{-\frac{r^2}{2}} r \, dr$ est une constante, donc :

$$
f_{\theta}(\theta) = \frac{1}{2\pi}, \quad \forall \theta \in [0,2\pi]
$$

Cela montre que $\theta$ suit une loi uniforme sur $[0,2\pi]$ :

$$
\theta \sim \mathcal{U}([0,2\pi])
$$

---

#### **Conclusion**
- $r^2$ suit une loi exponentielle $\mathcal{E}(1/2)$.
- $\theta$ suit une loi uniforme $\mathcal{U}([0,2\pi])$.

Ainsi, la distribution est isotrope, ce qui est coh√©rent avec la sym√©trie du probl√®me.

#### **3. M√©thode de g√©n√©ration de deux variables al√©atoires ind√©pendantes et de loi N (0,1)**

Nous avons montr√© pr√©c√©demment que si $X_1$ et $X_2$ suivent une loi normale centr√©e r√©duite $ \mathcal{N}(0,1) $, alors en passant aux coordonn√©es polaires, la variable $r^2$ suit une loi exponentielle $ \mathcal{E}(1/2) $ et la variable $ \theta $ suit une loi uniforme $ \mathcal{U}([0, 2\pi]) $.

La m√©thode de g√©n√©ration de deux variables al√©atoires ind√©pendantes suivant une loi normale centr√©e r√©duite $ \mathcal{N}(0,1) $ √† partir de ces r√©sultats est la suivante :

1. **G√©n√©rer $r$ et $ \theta $ :**
   - G√©n√©rer une variable al√©atoire $ U_1 \sim \mathcal{E}(1/2) $, c'est-√†-dire $ r^2 \sim \mathcal{E}(1/2) $ et donc $ r = \sqrt{U_1} $.
   - G√©n√©rer une variable al√©atoire $ U_2 \sim \mathcal{U}([0, 2\pi]) $, c'est-√†-dire que $ \theta \sim \mathcal{U}([0, 2\pi]) $.

2. **Reconstruction des variables $ X_1 $ et $ X_2 $ :**
   - Utiliser les coordonn√©es polaires pour obtenir $ X_1 $ et $ X_2 $ :
     $$
     X_1 = r \cos(\theta), \quad X_2 = r \sin(\theta)
     $$

Cela permet de g√©n√©rer deux variables al√©atoires ind√©pendantes $ X_1 $ et $ X_2 $ suivant une loi normale centr√©e r√©duite $ \mathcal{N}(0,1) $.

#### Conclusion

La m√©thode de g√©n√©ration est la suivante :
1. G√©n√©rer $ r \sim \mathcal{E}(1/2) $ et $ \theta \sim \mathcal{U}([0, 2\pi]) $.
2. Calculer $ X_1 = r \cos(\theta) $ et $ X_2 = r \sin(\theta) $.

Cela permet de cr√©er deux variables al√©atoires ind√©pendantes de loi $ \mathcal{N}(0,1) $.

#### **4. Transformation lin√©aire pour obtenir une variable al√©atoire Y ~ N(mu, sigma¬≤)**

Soit $ X \sim \mathcal{N}(0, 1) $, une variable al√©atoire suivant une loi normale centr√©e r√©duite. Pour obtenir une variable al√©atoire $ Y \sim \mathcal{N}(\mu, \sigma^2) $, il suffit d'appliquer la transformation lin√©aire suivante :

$$
Y = \mu + \sigma X
$$

Cette transformation permet de g√©n√©rer une variable $ Y $ suivant une loi normale de moyenne $\mu$ et de variance $\sigma^2$, √† partir d'une variable $ X $ suivant une loi normale centr√©e r√©duite.

#### **G√©n√©ration de 10 000 r√©alisations ind√©pendantes de la loi N(Œº=5, œÉ¬≤=4)**

La m√©thode de **Box-Muller** permet de g√©n√©rer des paires de variables al√©atoires ind√©pendantes suivant une loi normale centr√©e r√©duite $\mathcal{N}(0, 1)$. En utilisant cette m√©thode, on peut ensuite appliquer la transformation lin√©aire pour obtenir des r√©alisations de la loi $ \mathcal{N}(\mu=5, \sigma^2=4) $.

#### **Code Python pour g√©n√©rer les r√©alisations**

Voici comment vous pouvez g√©n√©rer $ T = 10000 $ r√©alisations ind√©pendantes de la loi $ \mathcal{N}(5, 4) $ en utilisant la m√©thode de Box-Muller et la transformation lin√©aire :


```python
import numpy as np
import matplotlib.pyplot as plt

# Param√®tres
mu = 5
sigma = 2  # √âcart-type (racine de la variance)
T = 10000  # Nombre de r√©alisations

# G√©n√©ration de paires de variables ind√©pendantes N(0,1) √† l'aide de la m√©thode de Box-Muller
u1 = np.random.rand(T // 2)
u2 = np.random.rand(T // 2)

z1 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
z2 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)

# Les variables z1 et z2 suivent une loi N(0,1)
# Transformation lin√©aire pour obtenir N(mu, sigma^2)
Y = mu + sigma * np.concatenate([z1, z2])

# Affichage des r√©sultats
plt.hist(Y, bins=30, density=True, alpha=0.6, color='g')
plt.title('Histogramme des 10 000 r√©alisations de N(5, 4)')
plt.xlabel('Valeur')
plt.ylabel('Densit√©')
plt.show()
```

### 1. Montrer que n√©cessairement $M \geq 1$

La m√©thode d'acceptation-rejet repose sur l'in√©galit√© suivante :

$$
f(x) \leq M g(x), \quad \forall x
$$

En divisant les deux membres par $g(x)$ (suppos√© non nul), on obtient :

$$
\frac{f(x)}{g(x)} \leq M, \quad \forall x
$$

Le coefficient $M$ est donc une borne sup√©rieure du rapport $\frac{f(x)}{g(x)}$. Par d√©finition, pour que la m√©thode fonctionne, il faut que cette borne soit suffisante pour garantir que l'in√©galit√© est respect√©e pour tout $x$. 

Or, ce rapport atteint n√©cessairement une valeur maximale quelque part, not√©e :

$$
M = \sup_{x} \frac{f(x)}{g(x)}
$$

Puisque $f(x)$ et $g(x)$ sont toutes deux des fonctions de densit√© de probabilit√©, elles sont normalis√©es √† 1 :

$$
\int f(x)dx = 1, \quad \int g(x)dx = 1
$$

Cela signifie qu'il existe au moins une valeur de $x$ pour laquelle $g(x)$ est plus grande que $f(x)$. En cons√©quence, la borne sup√©rieure de $\frac{f(x)}{g(x)}$ est n√©cessairement sup√©rieure ou √©gale √† 1 :

$$
M \geq 1
$$

### Conclusion :

La constante $M$ doit √™tre au moins √©gale √† 1 afin que la condition $f(x) \leq M g(x)$ soit respect√©e pour toute valeur de $x$, garantissant ainsi la validit√© de la m√©thode d'acceptation-rejet.

### 2. Peut-on g√©n√©rer une loi de Cauchy $C(0,1)$ √† partir d'une loi normale $N(0,1)$ ?

La densit√© de la loi de Cauchy $C(0,1)$ est donn√©e par :

$$
f(x) = \frac{1}{\pi(1 + x^2)}
$$

La densit√© de la loi normale $N(0,1)$ est donn√©e par :

$$
g(x) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{1}{2} x^2\right)
$$

La m√©thode d'acceptation-rejet n√©cessite qu'il existe une constante $M \geq 1$ telle que :

$$
f(x) \leq M g(x), \quad \forall x
$$

#### Analyse du rapport $\frac{f(x)}{g(x)}$

Calculons le rapport des deux densit√©s :

$$
\frac{f(x)}{g(x)} = \frac{\frac{1}{\pi(1 + x^2)}}{\frac{1}{\sqrt{2\pi}} \exp\left(-\frac{1}{2} x^2\right)}
$$

$$
= \frac{\sqrt{2\pi}}{\pi(1 + x^2)} \exp\left(\frac{1}{2} x^2\right)
$$

Ce rapport tend vers l'infini lorsque $|x| \to \infty$, en raison du terme exponentiel $\exp(x^2 / 2)$ qui cro√Æt beaucoup plus rapidement que le polyn√¥me $1 + x^2$.  

#### Conclusion

Puisque le rapport $\frac{f(x)}{g(x)}$ n'est pas born√©, il est impossible de trouver une constante $M$ finie telle que :

$$
f(x) \leq M g(x), \quad \forall x
$$

Cela signifie que la m√©thode d'acceptation-rejet ne peut pas √™tre appliqu√©e pour g√©n√©rer une loi de Cauchy $C(0,1)$ en utilisant une loi normale $N(0,1)$.

### 3- (a) Justification de l'√©galit√© :

$$
P(Y \leq y) = \sum_{i=0}^{+\infty} P(X \leq y, U \leq \frac{f(X)}{M g(X)}) \cdot \left[P(U \geq \frac{f(X)}{M g(X)})\right]^i
$$

#### Explication d√©taill√©e :

La variable $ Y $ est g√©n√©r√©e √† l'aide de la m√©thode d'acceptation-rejet. √Ä chaque it√©ration, une variable $ X \sim g(x) $ est g√©n√©r√©e, suivie de la g√©n√©ration d'une variable $ U $ uniforme sur $ [0, 1] $. Si la condition d'acceptation $ U \leq \frac{f(X)}{M g(X)} $ est satisfaite, alors $ X $ est accept√© comme r√©alisation de la loi $ f(x) $, c'est-√†-dire que $ Y = X $.

Sinon, si cette condition n'est pas remplie, $ X $ est rejet√© et un nouvel essai est effectu√©. Le processus peut √™tre vu comme un encha√Ænement d'√©checs et de r√©ussites, o√π √† chaque √©chec, la probabilit√© d'acceptation est fonction de la densit√© cible $ f(x) $ et de la densit√© auxiliaire $ g(x) $.

#### 1. Cas o√π l'on accepte du premier coup :

Dans ce cas, on g√©n√®re directement un $ X \leq y $ et la condition d'acceptation est satisfaite, c'est-√†-dire que $ U \leq \frac{f(X)}{M g(X)} $. La probabilit√© de ce premier succ√®s est donc donn√©e par :

$$
P(X \leq y, U \leq \frac{f(X)}{M g(X)})
$$

Cette probabilit√© repr√©sente la situation o√π $ X $ est inf√©rieur ou √©gal √† $ y $, et o√π $ U $ satisfait la condition d'acceptation. Cela correspond √† un succ√®s imm√©diat.

#### 2. Cas o√π l'on √©choue $ i $ fois avant d'accepter :

Si on √©choue $ i $ fois avant de r√©ussir, cela signifie qu'√† chaque it√©ration, la condition d'acceptation $ U \leq \frac{f(X)}{M g(X)} $ n'est pas remplie, donc on rejette chaque valeur de $ X $. La probabilit√© de rejet √† chaque essai est :

$$
P(U \geq \frac{f(X)}{M g(X)})
$$

On rejette $ X $ et on r√©p√®te le processus jusqu'√† ce que la condition d'acceptation soit remplie. Si on √©choue $ i $ fois, cela signifie que la condition de rejet a √©t√© satisfaite $ i $ fois cons√©cutivement. Cette probabilit√© de rejet se r√©p√®te sur $ i $ essais.

#### 3. Somme infinie des essais successifs :

Le nombre d'essais avant d'accepter une valeur suit une **loi g√©om√©trique**. En effet, √† chaque essai, on a une probabilit√© constante de succ√®s (d'acceptation) donn√©e par $ P(U \leq \frac{f(X)}{M g(X)}) $, et la probabilit√© d'√©chec est donn√©e par $ P(U > \frac{f(X)}{M g(X)}) $.

La probabilit√© totale d'accepter $ Y \leq y $ est donc la somme des probabilit√©s de succ√®s apr√®s $ i $ √©checs, pour chaque valeur possible de $ i $. Cette probabilit√© est la somme infinie des probabilit√©s de r√©ussite apr√®s $ i $ essais :

$$
P(Y \leq y) = \sum_{i=0}^{+\infty} P(X \leq y, U \leq \frac{f(X)}{M g(X)}) \cdot \left[P(U \geq \frac{f(X)}{M g(X)})\right]^i
$$

#### Conclusion :

Cette formulation montre que la probabilit√© d'accepter une variable $ Y \leq y $ suit bien la distribution cible $ f(x) $ apr√®s un nombre quelconque de rejets. L'√©galit√© est correcte et d√©crit le comportement du processus d'acceptation-rejet o√π la probabilit√© de succ√®s d√©pend de la condition $ U \leq \frac{f(X)}{M g(X)} $ et du nombre d'√©checs pr√©c√©dents. La somme infinie refl√®te les √©checs successifs avant d'accepter un $ X $.

### (b) Montrer que 
$$
P(X \leq y, U \leq \frac{f(X)}{M g(X)}) = \frac{1}{M} F_f(y)
$$
en notant $ F_f(y) = \int_{-\infty}^{y} f(t) \, dt $ la fonction de r√©partition de la loi de densit√© $ f(x) $.

#### Explication d√©taill√©e :

1. **Expression de $ P(X \leq y, U \leq \frac{f(X)}{M g(X)}) $** :

   On veut calculer la probabilit√© que la variable $ X $ soit inf√©rieure ou √©gale √† $ y $ et que la variable $ U $ soit inf√©rieure ou √©gale √† la fonction d'acceptation $ \frac{f(X)}{M g(X)} $. Cette probabilit√© est donn√©e par l'int√©grale conjointe des √©v√©nements $ X \leq y $ et $ U \leq \frac{f(X)}{M g(X)} $. Autrement dit, cela revient √† calculer la probabilit√© que l'√©chantillon $ X $ provenant de la densit√© auxiliaire $ g(x) $ soit dans l'intervalle $ (-\infty, y] $, et que le tirage de la variable $ U $ soit tel que $ U \leq \frac{f(X)}{M g(X)} $.

   Cette probabilit√© peut √™tre √©crite comme suit :

   $$
   P(X \leq y, U \leq \frac{f(X)}{M g(X)}) = \int_{-\infty}^{y} P\left( U \leq \frac{f(x)}{M g(x)} \Big| X = x \right) g(x) \, dx
   $$

   Ici, $ P\left( U \leq \frac{f(x)}{M g(x)} \Big| X = x \right) $ est la probabilit√© que $ U $ soit inf√©rieur ou √©gal √† la fonction d'acceptation donn√©e $ \frac{f(x)}{M g(x)} $, √©tant donn√© que $ X = x $. Puisque $ U $ suit une loi uniforme sur $ [0, 1] $, cette probabilit√© est simplement $ \frac{f(x)}{M g(x)} $, √† condition que $ \frac{f(x)}{M g(x)} \leq 1 $.

   Ainsi, l'expression devient :

   $$
   P(X \leq y, U \leq \frac{f(X)}{M g(X)}) = \int_{-\infty}^{y} \frac{f(x)}{M g(x)} g(x) \, dx
   $$

2. **Simplification de l'int√©grale** :

   L'int√©grale devient alors :

   $$
   P(X \leq y, U \leq \frac{f(X)}{M g(X)}) = \frac{1}{M} \int_{-\infty}^{y} f(x) \, dx
   $$

   Par d√©finition, $ \int_{-\infty}^{y} f(x) \, dx $ est simplement la fonction de r√©partition $ F_f(y) $ de la loi de densit√© $ f(x) $. On peut donc √©crire :

   $$
   P(X \leq y, U \leq \frac{f(X)}{M g(X)}) = \frac{1}{M} F_f(y)
   $$

#### Conclusion :

Nous avons montr√© que la probabilit√© de l'√©v√©nement o√π $ X \leq y $ et $ U \leq \frac{f(X)}{M g(X)} $ est √©gale √† $ \frac{1}{M} $ fois la fonction de r√©partition de la densit√© cible $ f(x) $ √† $ y $. Ce r√©sultat est essentiel pour d√©montrer que la m√©thode d'acceptation-rejet permet effectivement de g√©n√©rer des variables al√©atoires suivant la loi de densit√© $ f(x) $.

---

### (c) D√©monstration de $ P(Y \leq y) = F_f(y) $

Nous avons √©tabli que :

$$
P(X \leq y, U \leq \frac{f(X)}{M g(X)}) = \frac{1}{M} F_f(y)
$$

D'apr√®s la partie (a), la probabilit√© $ P(Y \leq y) $ s'exprime sous la forme :

$$
P(Y \leq y) = \sum_{i=0}^{+\infty} P(X \leq y, U \leq \frac{f(X)}{M g(X)}) \cdot \left[P(U \geq \frac{f(X)}{M g(X)})\right]^i
$$

---

#### √âtape 1 : Interpr√©tation comme une somme g√©om√©trique

On sait que :

$$
P(U \geq \frac{f(X)}{M g(X)}) = 1 - P(U \leq \frac{f(X)}{M g(X)})
$$

En utilisant le r√©sultat pr√©c√©dent :

$$
P(U \geq \frac{f(X)}{M g(X)}) = 1 - \frac{f(X)}{M g(X)}
$$

Ainsi, la somme infinie devient :

$$
P(Y \leq y) = \sum_{i=0}^{+\infty} \frac{1}{M} F_f(y) \cdot \left(1 - \frac{1}{M} F_f(y)\right)^i
$$

Nous reconnaissons ici une **s√©rie g√©om√©trique** de raison $ r = 1 - \frac{1}{M} $, qui converge sous la forme :

$$
\sum_{i=0}^{+\infty} a r^i = \frac{a}{1 - r}
$$

avec $ a = \frac{1}{M} F_f(y) $ et $ r = 1 - \frac{1}{M} $, donc :

$$
P(Y \leq y) = \frac{\frac{1}{M} F_f(y)}{1 - \left(1 - \frac{1}{M} \right)}
$$

---

#### √âtape 2 : Simplification

Dans le d√©nominateur :

$$
1 - \left(1 - \frac{1}{M} \right) = \frac{1}{M}
$$

Ainsi :

$$
P(Y \leq y) = \frac{\frac{1}{M} F_f(y)}{\frac{1}{M}}
$$

Finalement, nous obtenons :

$$
P(Y \leq y) = F_f(y)
$$

---

### Conclusion

Nous avons d√©montr√© que la m√©thode d'acceptation-rejet g√©n√®re une variable $ Y $ qui suit bien la loi de densit√© $ f(x) $, car sa fonction de r√©partition est exactement $ F_f(y) $, ce qui prouve que $ Y $ suit bien la loi de densit√© $ f(x) $.

Ainsi :

$$
P(Y \leq y) = F_f(y)
$$

### 4. Probabilit√© d'acceptation en un passage et nombre moyen de passages

#### Probabilit√© d'acceptation en un passage

L'algorithme d'acceptation-rejet accepte une valeur $ X $ si et seulement si la variable uniforme $ U $ satisfait la condition :

$$
U \leq \frac{f(X)}{M g(X)}
$$

La probabilit√© d'accepter $ X $ en un seul essai est donc donn√©e par l'esp√©rance conditionnelle :

$$
P(\text{acceptation}) = \mathbb{E} \left[ P\left(U \leq \frac{f(X)}{M g(X)} \mid X \right) \right]
$$

Puisque $ U $ est une variable uniforme sur $ [0,1] $, la probabilit√© que $ U $ soit inf√©rieure √† $ f(X) / M g(X) $ est simplement cette valeur :

$$
P(\text{acceptation}) = \mathbb{E} \left[ \frac{f(X)}{M g(X)} \right]
$$

En utilisant la d√©finition de l'esp√©rance :

$$
P(\text{acceptation}) = \int_{-\infty}^{+\infty} \frac{f(x)}{M g(x)} g(x) dx
$$

Or, par d√©finition de la densit√© de probabilit√© :

$$
\int_{-\infty}^{+\infty} f(x) dx = 1
$$
